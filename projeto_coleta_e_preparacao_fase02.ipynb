### Outliers

Optou-se por analisar os possíveis outliers de cada coluna numérica separadamente, utilizando um gráfico boxplot.






A coluna Q17 é numérica; no entanto, não será analisada, pois contém apenas os anos em que a pesquisa foi realizada e foi adicionada apenas para esse propósito.

#### Q8

```python
df['Q8'].plot(kind='box')
```


Foram observados outliers na coluna Q8, possivelmente decorrentes das correções nos dados feitas em 2016. Inicialmente, o formulário era aberto, mas posteriormente foi ajustado para uma escala de 0 a 10.


Devido a esses motivos, optou-se por remover os outliers utilizando como limite 1,5 desvios-padrão acima da média.

```python
# Média, desvio padrão e limites

mean = df['Q8'].mean()
std_dev = df['Q8'].std()

limite_superior = mean + 1.5 * std_dev
limite_inferior = mean - 1.5 * std_dev

print(limite_superior)
print(limite_inferior)
print(mean)
```

```python
# Verificando os outliers
outliers = df[(df['Q8'] > limite_superior) | (df['Q8'] < limite_inferior)]
print(outliers)
```

```python
# Retirando os outliers

df.loc[df['Q8'] > limite_superior, 'Q8'] = mean
df.loc[df['Q8'] < limite_inferior, 'Q8'] = mean
```

```python
df['Q8'].plot(kind='box')
```

O novo gráfico boxplot mostra que não há mais valores com outliers.







#### Q11

```python
df['Q11'].plot(kind='box')
```


Na coluna Q11, há muitos valores que são outliers, especialmente por se tratar da idade, podendo ser influenciados por diversos fatores que contribuem para a presença desses valores discrepantes.


Foi escolhida a **"regra dos 1,5*IQR"** para retirar os outliers devido à sua simplicidade e robustez. Essa técnica equilibra bem entre ser cuidadosa e sensível na detecção desses valores diferentes.

```python
Q1 = df['Q11'].quantile(0.25) #primeiro quartil
Q3 = df['Q11'].quantile(0.75) #terceiro quartil

IQR = Q3-Q1 #  intervalo interquartil

inferior = Q3+1.5 *IQR #limite inferior
superior = Q1-1.5 *IQR #limite superior

print(f'IQR = {IQR}')
print(f'Limite inferior = {inferior}')
print(f'Limite superior = {superior}')
```

```python
# Filtra a linha mantendo apenas as linhas onde a condição especificada
df = df[(df['Q11'] >= superior) & (df['Q11'] <= inferior)]
```
